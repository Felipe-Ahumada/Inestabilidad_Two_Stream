```python
import numpy as np
import matplotlib.pyplot as plt
import random as rm
plt.style.use('bmh')
```

# Condiciones iniciales

```python
#------------------------------------------------------------------------------------------------
#Condiciones iniciales


L = 6 #longitud de la línea, L no puede ser m'as pequeño que Dx
Dx = 6 #largo de cada intervalo, por tanto L / Dx casillas
Np = 10000 #número de cargas
Nc = L / Dx #número de celdas

#genera una posición (es necesaria multiplicarla por L) aleatoria de cada partícula
x = np.random.rand(Np) * L
v = np.random.rand(Np) * L
#x0.sort() #para ordenar la lista


linea = np.zeros(len(x))

#Si aumento la longitud de la línea y disminuyo el largo de intervalos hay una mayor oscilación
#Si disminuyo la longitud de la línea y aumento el largo de intervalos hay una menor oscilación
#parece que no es así xd

#------------------------------------------------------------------------------------------------

#Condiguración del tiempo
tmax = 100.0

h = 0.05

t = np.arange(0, tmax, h) 
```

# Condición de borde

En este modelo serán utilizadas condiciones de borde periódicas, esto se realiza para solucionar el caso en el que una partícula se escapa de los bordes delimitados en nuestra cuadrícula.

```python
def cond_borde(x, Dx):
    for i in range(len(x)):
        #así se soluciona el problema de que estén afuera de la línea
        if x[i] > L:
            x[i] += -L #si se sale por L, entonces aparece en 0
            
        if x[i] < 0:
            x[i] += L #si se sale por 0, entonces aparece en L
```

# Densidad de partículas

```python
def n(x, Dx=6, Nc=1.0):
    
    n0 = np.floor(x / Dx)
    
    rho = np.zeros(int(Nc)) # densidad de cargas por celda

    #aquí contamos cuántos electrones hay en cada celda
    for s in range(len(n0)):

        rho[int(n0[s])] += 1
        
    return 1 - rho * ( Nc / len(x) ) #Nc cantidad de celdas, len(x) = Np; con esto se normaliza la densidad
```

```python
Dx
```

# Campo eléctrico

```python
def E(rho, Dx=6):
    
    e = Dx * np.cumsum(rho)
    
    return e - np.mean(e)
```

# Particle pusher

```python
def push(x, v, Et, Dt=h):
    
    c = np.floor(x / Dx).astype(int)
    
    v = v - Dt * Et[c]
    
    x = x + Dt * v
    
    return x, v
```

# Guardado de posiciones y velocidades

```python
#------------------------------------------------------------------------------------------------

xx = np.empty((t.size, Np))
vv = np.empty((t.size, Np))

#------------------------------------------------------------------------------------------------

for tau in range(t.size):
        
    #Condiciones iniciales
    cond_borde(x, Dx)
    
    RHO = n(x)
    
    #array del campo electrico
    Et = E(RHO)
    
    #guarda las nuevas posiciones y velocidades
    xx[tau] = x
    vv[tau] = v
    
    #actualiza las posiciones y velocidades
    x, v = push(x, v, Et)
```

# Resultado

```python
plt.plot(t, xx)
plt.xlabel('tiempo[s]')
plt.ylabel('posición')
```

```python
def plotfunct(t):
    return plt.plot(xx[t],linea, '.'), plt.xlim(0 - L, 2*L)
```
